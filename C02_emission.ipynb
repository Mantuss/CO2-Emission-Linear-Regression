This data set contains information on how much CO2 a vehicle emits and how it varies by different feature. There are 7385 rows and 12 columns in this dataset.

**Columns in the Dataset(Descriptions):**

    1) Make = Specifies the car branding(Company).
    2) Model = Specifies the Vehicle's model.
    3) Vehicle Class = Specifies the Vehicle's class.
    4) Engine Size(L) = Specifies engine size based on fuel capacity. 
    5) Cylinders = Number of cylinder in the vehicle.
    6) Transsiom = Specifies the transmission of the vehicle(It's what moves the power from the engine to the wheels).
    7) Fuel Type = Specifies the fuel elligible for the vehicle.
    8) Fuel Consumption city = Fuel Consumption in city Roads. 
    9) Fuel Consumption Hwy = Fuel Consumption in HighWay roads.
    10) Fuel Consumption Comb = Fuel Consumption in both Highway and City roads.
    11) Fuel Consumption Comb(mpg) = Fuel consumption Combined in meter per gallon.
    12) CO2 Emissions(g/km) = The total co2(Carbondioxide) Emitted by the vehicle.
